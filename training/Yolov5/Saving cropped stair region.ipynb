```python
d = {}
for i in range(1329,1696):
  try: 
    img = f"./new_dataset/{i}.png"
    results = model(img)
    cropped_img = results.crop()[0]['im']
    cv2.imwrite(f"./cropped_images/cropped{i}.png", cropped_img)
    l = results.xyxy[0]
    d[i] = {"xmin": int(l[0,0]),
            "ymin": int(l[0,1]),
            "xmax": int(l[0,2]),
            "ymax": int(l[0,3]),
            "conf": round(float(l[0,4]), 2)}
    print(i, d[i])
  except: 
    pass
```

```python
with open("data.json", 'w') as f:
  json.dump(d, f)
```
